```python
import matplotlib.pyplot as plt
```

```python
total_lines = 6511404
tweets_right_now = 5470346

# Example data
values = [total_lines, tweets_right_now]  # Two integer values
labels = ["Number of possible tweets", "Number of stored tweets"]  # Corresponding labels for the bars
# Creating the bar chart
plt.bar(labels, values)  # You can choose colors

# Adding title and labels
plt.title('Comparison of tweets provided vs stored')
plt.ylabel('Number of tweets')
```
