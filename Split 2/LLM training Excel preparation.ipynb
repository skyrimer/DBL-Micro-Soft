```python
import json
import pandas as pd
import openpyxl

# Function to read the JSON file and extract the "text" field
def extract_text_from_json(json_file_path):
    texts = []
    with open(json_file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                data = json.loads(line)
                text = data.get('text') if data.get('full_text') == None else data.get('full_text')
                if text:
                    texts.append(text)
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON: {e}")
    return texts

# Function to write the extracted data to an Excel file
def write_to_excel(data, excel_file_path):
    df = pd.DataFrame(data, columns=['text'])
    df.to_excel(excel_file_path, index=False)

# Main function
def main():
    json_file_path = 'c:/Users/20232075/Downloads/data/airlines-1558527599826.json'  # Replace with your JSON file path
    excel_file_path = 'C:/Users/20232075/Downloads/aboba_pisyapopa.xlsx'  # Replace with your desired Excel file path

    texts = extract_text_from_json(json_file_path)
    write_to_excel(texts, excel_file_path)
    print(f"Data has been successfully written to {excel_file_path}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from googletrans import Translator
from deep_translator import GoogleTranslator
import time

def translate_if_needed(text):
    try:
        # Detect and translate the text to English if it is not in English or Russian
        translation = GoogleTranslator(source='auto', target='en').translate(text)
        return translation
    except Exception as e:
        print(f"Error translating text: {text}\nError: {e}")
        return text

# Function to process the Excel file
def process_excel_file(input_file_path, output_file_path):
    df = pd.read_excel(input_file_path)
    
    # Apply translation function to the 'text' column
    df['translation'] = df['text'].apply(translate_if_needed)
    
    # Save the updated DataFrame back to an Excel file
    df.to_excel(output_file_path, index=False)
    print(f"Data has been successfully written to {output_file_path}")

def main():
    input_file_path = 'C:/Users/20232075/Downloads/aboba_pisyapopa.xlsx'  # Input Excel file path
    output_file_path = 'C:/Users/20232075/Downloads/translated.xlsx'  # Output Excel file path

    process_excel_file(input_file_path, output_file_path)

if __name__ == "__main__":
    main()
```
